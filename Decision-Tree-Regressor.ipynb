```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first = True)
```

```python
dataset
```

```python
Independent = dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
```

```python
Independent
```

```python
Dependent = dataset[['Profit']]
```

```python
Dependent
```

```python
X_train,X_test,y_train,y_test = train_test_split(Independent,Dependent,test_size=0.40,random_state=0)
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)

X_test = sc.transform(X_test)
```

```python
X_train
```

```python
regressor = DecisionTreeRegressor(min_samples_split=4, splitter='best',criterion='squared_error', random_state=0 )
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

```python
filename = "Final_model_DecisionTreeRegressor.pkl"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open('Final_model_DecisionTreeRegressor.pkl','rb'))
```

```python
R_D_spend = int(input("Enter the research and development spending value"))
Administration_spend = int(input("Enter the Administraion spending value"))
Marketing = int(input("Enter the Marketing spending value"))
result = loaded_model.predict([[R_D_spend,Administration_spend,Marketing,0,0]])
print("To predicted the total profit based on your spending amount in each deparment is", result)
```
